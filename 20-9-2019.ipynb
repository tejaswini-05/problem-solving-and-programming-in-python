## Tasks

* Function to print all numbers divisible by 6 and not a factor of 100 in a given range(lb, ub) inclusive

```python
def div(lb,ub):
    for i in range(lb,ub+1):
        if i%6==0 and 100%i!=0:
            print(i)
div(1,100)            
```

* Function to find the average of cubes of all the even numbers in a given range(lb, ub) inclusive

```python
def av(lb,ub):
    res=0
    count=0
    for i in range(lb,ub+1):
        if i%2==0:
            res=res+i**3
            count=count+1
    print(int(res/count))
av(1,10)
```

* Function to generate the list of factors for a given number

```python
def val(lb,ub):
    for i in range(lb,ub+1):
        if ub%i==0:
            print("\n",i)
val(int(input()),int(input()))  
```

* Function to calculate the factorial of a given number

```python
def fact(a):
    facto=1
    for i in range(1,a+1):
        facto=facto*i
    print(facto)  
fact(int(input()))    
```

## advanced problem
* Function to calculate the average of all factorials in a given range

```python
def avfac(a):
    fac=1
    out=0
    count=0
    for i in range(1,a+1):
        fac=fac*i
        out=out+fac
        count=count+1
    print("\n",out/count)
avfac(int(input()))
```

* Function to generate N odd armstrong numbers

```python
def arm(i):
    r=0
    m=i
    while i>0:
        rem=i%10   
        r=r+rem**3
        i//=10
    if m==r:
         print("\n",r)    
arm(int(input()))   
      
```

```python
def odd(lb,ub):
    r=0
    for i in range(lb,ub):
        m=i
        print(i)
        while m>0:
                rem=m%10  
                print(rem) 
                r=r+rem**3
                m//=10
        if i==r:
                if r%2==0:
                    print(r)
odd(1,1000)
```

```python
# odd numbers
def od(lb,ub):
    for i in range(lb,ub+1):
        if i%2!=0:
            print(i)
od(1,10)            
```

* Function to generate Multiplication table for a number in a given range

```python
def multi(n,l):
    for i in range(l,l+3):
        print(n,'x',i,'=',n*i)
multi(int(input()),int(input()))        
        
```
